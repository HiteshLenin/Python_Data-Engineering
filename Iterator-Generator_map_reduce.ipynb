```python
def is_prime(num):
    '''
    Naive method of checking for primes. 
    '''
    for n in range(2,num):
        if num % n == 0:
            print('not prime')
            break
    else: # If never mod zero, then prime
        print('prime')
```

```python
is_prime(31)
```

```python
## Iterators and Generators
# Iterators
nums = [10, 20, 30]

it = iter(nums)

print(next(it))  # 10
print(next(it))  # 20
print(next(it))  # 30
```

```python
for x in nums:
    print(x)
```

```python
#Generators
#A generator is a special type of iterator that is written like a function but uses the yield keyword instead of return.
#It pauses when it yields, and resumes from where it left off.
# Lazy Evaluation
def gencubes(n):
    for num in range(n):
        yield num**3
```

```python
gencubes(15)
```

```python
for x in gencubes(15):
 print(x)
```

```python
def genfibon(n):

    a = 1
    b = 1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
genfibon(6)
```

```python
for x in genfibon(9):
    print(x)
```

```python
# map()

#The map() is a function that takes in two arguments: 
#1. A function 
#2. A sequence iterable.
#map() applies a function to each element of an iterable and returns a map object (iterator).
# it does not returns a list, but an iterator
#map() is lazy evaluated
#Values are computed only when iterated
def square(x):
    return x * x

nums = [1, 2, 3]

print(list(map(square, nums)))


```

```python
nums = [1, 2, 3, 4]

result = map(lambda x: x * 2, nums)

print(result)           # <map object at 0x...>
print(list(result))     # [2, 4, 6, 8]
```

```python
 #reduce()
#reduce() takes many values and reduces them to ONE value.
#If seq = [s1, s2, s3, ... , sn], calling reduce(function, sequence) works like this:
#At first the first two elements of sequence will be applied to function, i.e. func(s1,s2)
#The list on which reduce() works looks like this: [ function(s1, s2), s3, ... , sn ]
#In the next step the function will be applied on the previous result and the third element of the list, i.e. function(function(s1, s2),s3)
#The list looks like: [ function(function(s1, s2),s3), ... , sn ]
```

```python
from functools import reduce
```

```python
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, nums)
print(result)
```

```python
reduce(lambda x, y: x * y, [1, 2, 3, 4])
```

```python
reduce(lambda x, y: x if x > y else y, [3, 7, 2, 9])
```

```python
# Filter
#filter() returns elements from an iterable for which the function returns True.
nums = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, nums)

print(list(result))
```

```python
words = ["hitesh", 'lenin', 'Michael','mark']
long_words = list(filter(lambda word: len(word)>=6, words))

print(long_words)
```
