```python
#Type Conversion in Python
##Explicit Type Conversion (Manual)
##manually convert one data type to another using built-in functions.
num = 100
text = str(num)

print(text)        
print(type(text))

value = "50"
number = int(value)
print(number)
##Implicit Type Conversion (Automatic)
##Python automatically converts one data type to another when it is safe.
```

```python
##Boolean Data Type
##A Boolean represents
# TRUE
# FALSE

x = 15
y = 18
print(x>=y)
print(x<=y)
print(x==y)
print(x != y)
```

```python
##Logical Operators
##AND#Returns True only if ALL conditions are True.
##OR#Returns True if AT LEAST ONE condition is True.
##NOT#Reverses the Boolean value.
a = True
b = False
print(a and b)  
print(a or b)  
print(not a)    
```

```python
#Operators in Python
a = 10
b = 3
print(a + b)  
print(a - b)
print(a * b)
print(a / b)
print(a % b) #Returns the remainder of division.
print(a ** b) #Raises a number to the power of another.
print(a // b)#Divides and returns the integer floor value.
```

```python
#f-strings (Formatted String Literals)
#Most readable and efficient way to format strings
#Variables are placed directly inside {}

name = "Hitesh"
age = 26

print(f"My name is {name} and my age is {age}")
```

```python
x = 10
y = 20

print(f"Sum is {x + y}")

```

```python
#.format() Method
#Older but still valid string-formatting method
#Uses {} placeholders
#Values passed via .format()
#Positional formatting
print("My name is {} and age is {}".format("Hitesh", 26))
```

```python
#Indexed formatting
print("Name: {0}, Age: {1}".format("Hitesh", 26))
```

```python
#Named placeholders
print("Name: {name}, Age: {age}".format(name="Hitesh", age=26))
```

```python
#if – elif – else Statement
marks = 80
if marks >= 90:
    print("Grade A")
elif marks >=75:
    print("Grade B")
elif marks >=60:
    print("Grade C")
else:
    print("Fail")
    
```

```python
#Nested if Statements ( if inside if )
#A nested if means placing one if statement inside another if block.
username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Incorrect password")
else:
    print("Invalid username")

```
