```python
# For Loop
#A for loop is used when the number of iterations is known or bounded.
#It iterates over iterables (list, tuple, set, dictionary, string, range, generator).
numbers = [1, 2, 3, 4]

for i in numbers:
    print(i)

    
```

```python
#While Loop
#A while loop keeps executing as long as the condition is TRUE.
#The moment the condition becomes FALSE, the loop stops.

n = 11
sum = 0
i = 1

while i <= n:
    print(i)
    i += 2
```

```python
#Break & Continue
#break immediately terminates the loop, even if the loop condition is still true.
for ch in "stnring":
    if ch == "n":
        break
    print(ch)

print("Loop ended")
```

```python
i = 1
while i <= 10:
    if i == 5:
        break
    print(i)
    i += 1
```

```python
#continue statement
#continue skips the current iteration and moves to the next iteration of the loop.
for ch in "string":
    if ch == "r":
        continue
    print(ch)
```

```python
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```
