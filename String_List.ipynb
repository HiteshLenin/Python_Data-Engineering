```python
#Prime number
#it is divided by one and itself
n = 5
is_prime = True

for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
        is_prime = False
        break

print("Prime" if is_prime and n > 1 else "Not Prime")

```

```python
#Factorial
n = 5
fact = 1
for i in range(1, n + 1):
    fact *= i
print(fact)
```

```python
#Palindrome Problems
#A value that reads the same forward and backward
s = "Hello World Python"
result = ""

for ch in s:
    if ch == " ":
        result += "_"
    else:
        result += ch.lower()

print(result)

```
